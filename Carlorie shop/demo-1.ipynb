```python
import pandas as pd
from datetime import datetime
```

```python
class run:
    _user_data=None
    _food_data=None
    _user=None
    _daily_cal=0
    _current_time = datetime.today().strftime('%Y-%m-%d')
    def __init__(self,user=None):
        self._current_time = datetime.today().strftime('%Y-%m-%d')
        self._user_data=pd.read_csv('USER_INFO\\USER_BASE_INFO.CSV') 
        self._food_data=pd.read_csv('USER_INFO\\USER_FOOD_LOGS.CSV')
        if user is not None:#  read the users info
            tar_info=self._user_data[self._user_data.name==user].sort_values(by="time").to_dict('index')
            tar_info=tar_info[list(tar_info.keys())[-1]]
            self._user=tar_info
            self._daily_cal=tar_info['daily_cal']

    def eat(self,foodname=None,calorie_rate=None,protein=None,carbohydrate=None,fat=None,weight=None,calcost=None,types=None,save_choice=None):        
        if (foodname is not None) and (foodname in self._food_data.food_name.to_list()):
            tar_info=self._food_data[self._food_data.food_name==foodname].to_dict('index')
            tar_info=tar_info[list(tar_info.keys())[-1]]
            foodname=tar_info['food_name']
            calorie_rate=tar_info['calorie_rate']
            protein=tar_info['protein']
            carbohydrate=tar_info['carbohydrate']
            fat=tar_info['fat']    

            if (types is None ) or types not in ['B','BA','L','LA','D','DA']:
                types=input("please input the meal types('B','BA','L','LA','D','DA'):")
            if weight==0:
                weight=input("Please input the weight:")
            
        else:
            if foodname is None:
                foodname=input("please input food name:")
            if types is None:
                types=input("please input the meal types('B','BA','L','LA','D','DA'):")
            if calorie_rate is None:
                calorie_rate=input("please input food calorie_rate(per 100g):")      
            if protein is None:
                protein=input("please input food protein(per 100g):")      
            if carbohydrate is None:
                carbohydrate=input("please input food carbohydrate(per 100g):")      
            if fat is None:
                fat=input("please input food fat(per 100g):")      
            if weight is None:
                weight=input("please input food weight:")
            if calcost is None:
                    weight=int(weight)
                    calorie_rate=int(calorie_rate)
                    calcost=weight*calorie_rate/100
        
        
        new_info=[len(self._food_data)+1,len(self._food_data)+1,foodname,self._user['name'],self._current_time,types,weight,calorie_rate,carbohydrate,protein,fat,calcost]
        ccc=self._food_data.columns.to_list()
        new_data=pd.DataFrame([new_info],columns=ccc)
        self._food_data=self._food_data.append(new_data)
        print(dict(zip(ccc,new_info)))

        if save_choice is None:
            save_choice=input("save?")
        if save_choice=='Y':
            self._food_data.to_csv('USER_INFO\\USER_FOOD_LOGS.CSV',index=False)    


    def read(self,date_log=None):
        if date_log is None:
            date_log=self._current_time
        self._food_data=pd.read_csv('USER_INFO\\USER_FOOD_LOGS.CSV')
        result=self._food_data
        result=result[(result.time==date_log)&(result.user_name==self._user['name'])]
        
        cal_cost=result['calorie_cost'].to_list()
        cal_left=self._daily_cal
        for i in range(len(cal_cost)):
            cal_left-=cal_cost[i]
        print(f"Todat's calorie left: {cal_left}")

        return result      
             
```
