```python
import pandas as pd
```

```python
# read foodfile
foodfile = pd.read_csv('food.csv', iterator = True) # iterator = True: read in chunk
loop = True
chunkSize = 200000
chunks = []
while loop:
    try:
        chunk = foodfile.get_chunk(chunkSize)
        chunks.append(chunk)
    except StopIteration:
        loop = False
        print('Done')

df = pd.concat(chunks, ignore_index=True)
```
